### Generating fake images – digits

```python
from functional import gen
import matplotlib.pyplot as plt
%matplotlib notebook
from numpy.random import RandomState
from nn import generate_noise

rng = RandomState(12333)

Z = generate_noise(rng, batch_size=128)
generated_images = gen(Z, *g_params).eval()

f, axarr = plt.subplots(4,5)

for i in range(4):
    for j in range(5):
        plt.axes(axarr[i,j])
        plt.xticks([])
        plt.yticks([])
        plt.imshow(generated_images[5*i+j].reshape((28,28)), cmap='gray')
```

### Smooth transitions on linear interpolations

```python
for i in range(5):
    for j in range(10):
        Z[10*i+j] = Z[10*i] + float(j)/9. * (Z[10*i+9] - Z[10*i])

generated_images = gen(Z, *g_params).eval()

f, axarr = plt.subplots(5,10)

for i in range(5):
    for j in range(10):
        plt.axes(axarr[i,j])
        plt.xticks([])
        plt.yticks([])
        plt.imshow(generated_images[10*i+j].reshape((28,28)), cmap='gray')
```
